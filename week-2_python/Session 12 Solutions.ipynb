### Basic Plotting with Matplotlib
#### Example 1: Line Plot
Instructions:

- Import the matplotlib.pyplot module.
- Create a simple line plot using `plt.plot()`.
- Display the plot using `plt.show()`.


```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Creating the line plot
plt.plot(x, y)

# Adding title and labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Displaying the plot
plt.show()
```

**Example 2: Customizing Line Plot**

Instructions:

- Customize the plot by adding color, line style, and markers.
- Use `plt.grid()` to add grid lines.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Creating the line plot with customization
plt.plot(x, y, color='green', linestyle='--', marker='o')

# Adding title and labels
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Adding grid lines
plt.grid(True)

# Displaying the plot
plt.show()
```

**Example 3: Bar Plot**

Instructions:

**Example 4: Histogram**

Instructions:

**Example 6: Creating Subplots**

Instructions:

- Create multiple plots in a single figure using `plt.subplot()`.
- Display the plots using `plt.show()`.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Creating subplots
plt.subplot(2, 1, 1)  # (rows, columns, panel number)
plt.plot(x, y1, 'r')
plt.title('First Plot')

plt.subplot(2, 1, 2)
plt.plot(x, y2, 'g')
plt.title('Second Plot')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

**Practice Problem 6: Create Your Own Subplots**

Create a figure with two subplots:

- A line plot for the data x: [0, 1, 2, 3, 4], y: [0, 1, 4, 9, 16]
- A bar plot for the data categories: ['A', 'B', 'C'], values: [5, 10, 15]

**Solution:**

**Adding Annotations**


Example 7: Adding Annotations

**Instructions:**
- Add annotations to a plot using `plt.annotate()`.
- Customize the annotations to highlight specific data points.


```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Creating the line plot
plt.plot(x, y, marker='o')

# Adding annotation
plt.annotate('Peak', xy=(5, 11), xytext=(4, 10),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Adding title and labels
plt.title("Plot with Annotation")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Displaying the plot
plt.show()
```

**Pie Chart:**

Create a pie chart for the following data: 
- Labels: ['A', 'B', 'C', 'D'] 
- Sizes: [15, 30, 45, 10].

**Solution:**


```python
import matplotlib.pyplot as plt

# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Creating the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Adding title
plt.title("Pie Chart")

# Displaying the plot
plt.show()
```
